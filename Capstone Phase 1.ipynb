```python
import csv
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.chrome.options import Options as ChromeOptions
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException

def initialize_driver()
    
    chromedriver_path = r"C:\Users\hp\Desktop\chromedriver2\chromedriver-win32\chromedriver.exe"
    
    chrome_options = ChromeOptions()
    

    driver_service = ChromeService(chromedriver_path)
    driver = webdriver.Chrome(service=driver_service, options=chrome_options)
    return driver


def close_popup(driver):
    try
        
        close_button = driver.find_element(By.XPATH, '//*[@class="bui-tab__link "]')
        close_button.click()
    except NoSuchElementException:
        pass


def save_to_csv(data, filename):
    try:
        with open(filename, "w", newline="", encoding="utf-8") as f:
            writer = csv.writer(f)
            writer.writerow(["Hotel Name", "Hotel URL", "Hotel Rating"])  # Write header
            for item in data:
                writer.writerow([item["hotel_name"], item["hotel_url"], item["hotel_rating"]])
    except Exception as e:
        print("An error occurred while saving to CSV:", str(e))


cities = ["Dubai", "London", "Kuala Lumpur", "Manchester", "New Delhi",
          "Birmingham", "Berlin", "Sydney", "Melbourne", "Paris", "Tokyo", "Toronto"]


csv_file_names = ["Data1.csv", "Data2.csv", "Data3.csv"]


driver = initialize_driver()


for city in cities:

    
    driver.get("https://www.booking.com/searchresults.html?ss=" + city)

    
    close_popup(driver)

    
    time.sleep(5)

    
    search_input = driver.find_element(By.XPATH, f"//[@id='search_results_table']/div[1]//[contains(@class, 'sb-searchbox__input')]")
    search_input.clear()  
    search_input.send_keys(city)  

    
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_element_located((By.XPATH, "//div[@data-testid='property-card']")))

    
    hotel_listings = []

    
    for hotel_box in driver.find_elements(By.XPATH, "//div[@data-testid='property-card']"):
        try:
            hotel_name = hotel_box.find_element(By.XPATH, ".//h3/a[@data-testid='title-link']").text
            hotel_url = hotel_box.find_element(By.XPATH, ".//h3/a[@data-testid='title-link']").get_attribute("href")
            hotel_rating = hotel_box.find_element(By.XPATH, ".//div[@data-testid='property-rating']").text

            
            hotel_listings.append({
                "hotel_name": hotel_name,
                "hotel_url": hotel_url,
                "hotel_rating": hotel_rating,
            })
        except NoSuchElementException:
            print("Hotel information not found for a listing. Skipping...")

    
    for i in range(len(csv_file_names)):
        save_to_csv(hotel_listings, csv_file_names[i])

driver.quit()
```
